```
from tensorflow.examples.tutorials.mnist import input_data
```

```
mnist = input_data.read_data_sets("MNIST_data", one_hot=True)
```

```
X = mnist.train.images
```

```
y = mnist.train.labels
```

```
import numpy as np
```

```
index = np.where((y[:, 1]!=0) | (y[:, 7]!=0))[0]
```

```
X_train = X[index, :]
X_train.shape
```

```
y_train = y[index, 1]
y_train.shape
```

```
from sklearn.svm import SVC, LinearSVC
```

```
from sklearn.model_selection import GridSearchCV
```

```
gs = GridSearchCV(LinearSVC(), param_grid={"C":[2**i for i in range(-15, 13, 1)]},
                 cv=5, return_train_score=True, n_jobs=-1)
```

```
gs.fit(X_train, y_train)
```

```
best_es = gs.best_estimator_
```

```
X_test = mnist.test.images
```

```
y_test = mnist.test.labels
```

```
index_test = np.where((y_test[:, 1]!=0) | (y_test[:, 7]!=0))[0]
```

```
X_te = X_test[index_test, :]
y_te = y_test[index_test, 1]
```

```
best_es.score(X_te, y_te)
```

```
train_sc = gs.cv_results_["mean_train_score"]
test_sc = gs.cv_results_["mean_test_score"]
```

```
train_err = 1- train_sc
test_err = 1- test_sc
```

```
plt.semilogx(np.array([2**i for i in range(-15, 13, 1)]), train_err, "b--", label="Train error")
plt.semilogx(np.array([2**i for i in range(-15, 13, 1)]), test_err, "r-", label="Test error")
plt.legend()
plt.xlabel("C")
plt.ylabel("Error")
plt.show()
```
