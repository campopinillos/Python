# Sesión 2: Scikit-Learn

Scikit-Learn es una librería en Python que provee un conjunto de herramientas para minería y análisis de datos. Scikit-Learn está construído sobre NumPy, SciPy y matplotlib. Además es código abierto que puede ser usable comercialmente. Scikit-Learn cuenta con 6 grandes módulos, cada uno de ellos especializado en diferentes tareas y algoritmos del esquema de trabajo en aprendizaje de máquina.

La funcionalidad de cada módulo está distribuida de la siguiente manera:

## Aprendizaje supervisado

<img src="https://drive.google.com/uc?export=view&id=1rjuuotQqxep8nKLwn5JPGSuvww-FYj4y">

### Clasificación
Cada algoritmo en Scikit-learn es expuesto a través de un objeto ''Estimator''. La ventaja es que todos los algoritmos de Scikit-learn tienen una interfaz muy consistente. En el módulo de clasificación, encontramos implementaciones de los siguientes algoritmos:
* Regresión logística
* Gradiente descendente estocástico
* Naïve Bayes
* Árboles de decisión
* Random Forest
* Máquinas de vectores de soporte

### Regresión
Entre los algoritmos implementados de regresión, encontramos:
* Regresión lineal
* Regresión usando vectores de soporte

## Aprendizaje no supervisado
Para aprendizaje no supervisado, encontramos un amplia gama de técnicas implementadas en Scikit-Learn.

### Clustering
<img src="https://drive.google.com/uc?export=view&id=1nF23xzJ-qtd7tt2Sth3lI9vzaqx7JsFO">

Scikit-Learn implemente varias técnicas de agrupamiento (Clustering) tales como:
* K-medias
* Clustering jerárquico
* Clustering espectral
* DBSCAN

Adicionalmente provee una serie de funcionalidades para medir el rendimiento de cada algoritmo:
* Homogeneidad, completitud y V-score
* Medidas usando información mutua
* Coeficiente silueta

### Reducción de la dimensionalidad
Scikit-Learn tambien provee una serie de técnicas eficientemente implementadas para reducción de la dimensionalidad:
* Análisis de componentes principales (PCA)
* Factorización de matrices no negativas

## Extracción de características y selección de modelos

### Selección de modelos
En este módulo se provee la funcionalidad necesaria para hacer comparar, validar y escoger los parámetros de un modelo:
* Validación cruzada
* Medidas de desempeño:
    * Exactitud
    * Matriz de confusión
    * Precisión, recuperación y F-score
    * Medidas para problemas multi-etiqueta
    * Medidas para regresión
* Exploración usando una grilla de parámetros

### Preprocesamiento
En este módulo se encuentra las implementaciones para extracción de características en texto e imágenes. También se pueden encontrar funcionalidades para preprocesar un conjunto de datos:

* Extracción de características en texto:
    * Bolsa de palabras
    * Tokenización
* Preprocesamiento:
    * Normalización
    * Escalamiento

# Estimator API

<img src="https://drive.google.com/uc?export=view&id=1nnjm7QQh-uIFcPpn0NjQ8EIbYR4Meji2">

Este es el flujo de trabajo para aprendizaje supervisado. Cada objetivo de tipo `Estimator` tiene funciones implementadas para cubrir cada parte de este flujo de trabajo:
* Entrenamiento: `model.fit(X_train, y_train)`
* Predicción: `model.predict(X_test)`
* Evaluación: `model.score(X_test, y_test)`
