**Crear un gráfico de área de todas las columnas pero sólo para las filas desde la 0 hasta la número 30. (Pista: use .iloc).**

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.stackplot(np.arange(0,31),df3['a'].iloc[:31],df3['b'].iloc[:31],
             df3['c'].iloc[:31],df3['d'].iloc[:31],
             labels=['a','b','c','d'], alpha=0.4,linestyle='solid')
ax.legend(loc='upper right')
plt.show()
```

## Pregunta Extra!

**Como se puede observar en la figura anterior, la leyenda queda por encima de una parte de la Figura. Por favor ubique la leyenda como se muestra a continuación.**

**Ver: http://stackoverflow.com/questions/23556153/how-to-put-legend-outside-the-plot-with-pandas**

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.stackplot(np.arange(0,31),df3['a'].iloc[:31],df3['b'].iloc[:31],
             df3['c'].iloc[:31],df3['d'].iloc[:31],
             labels=['a','b','c','d'], alpha=0.4,linestyle='solid')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
